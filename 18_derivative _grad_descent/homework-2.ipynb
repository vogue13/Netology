### Задание 4:

Градиентный спуск своими руками

```python
def func(x):
    return (x[0]**2) + (x[1]**2) 
```

```python
x = np.array([100, 200])
lr = 0.1
```

```python
i=0

eps = np.sqrt(np.finfo(float).eps)

for i in range(100):
    grad = approx_fprime(x, func, [eps, eps])
    x = x - lr*grad
    i += 1
```

```python
print("Минимум достигается в: ", x)
print("Значение функции в минимуме: ", func(x))
```

Проверка с помощью встроенной функции scipy: 

```python
res = minimize(func, x, method='nelder-mead', options={'xtol': 1e-8, 'disp': True})
```

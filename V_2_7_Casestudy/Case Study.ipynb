
### 1) Как критики относятся к спортивным играм?

H0: средняя удовлетворенность критиков спортивными играми >= 50 баллов

H1: средняя удовлетворенность критиков спортивными играми < 50 баллов

За alpha здесь и далее будем принимать 0.05

Сравниваем значение по выборке с константой, поэтому нужен одновыборочный критерий.

Тест нужен односторонний, т.к. нас интересует только неравенство в меньшую сторону.

```python
df = pd.read_csv('video_games_sales.csv')
df
```

```python
df1 = df[df['Genre'] == 'Sports']['Critic_Score'].dropna()
```

```python

alpha = 0.05
result = st.ttest_1samp(df1, 50)
print(result)

if (result.statistic < 0) & (result.pvalue / 2 < alpha):
    print('Отвергаем нулевую гипотезу, среднее меньше 50 баллов')
else:
    print('Не отвергаем нулевую нулевую гипотезу, средняя удовлетворенность критиков спортивными играми >= 50 баллов')
```

Значение t-теста существенно выше 0, соответственно, средняя выборки больше значения, которое нам необходимо, чтобы не отвергнуть нулевую гипотезу.

### 2) Критикам нравятся больше игры на PC или на PS4?

H0: критикам одинаково нравятся PC и PS4 

H1:  критикам больше нравится PC 

Используем двухвыборочный двухсторонний критерий.  


```python
df2 = df[df['Platform'] == 'PC']['Critic_Score'].dropna()
df3 =df[df['Platform'] == 'PS4']['Critic_Score'].dropna()
```

```python
result = st.ttest_ind(df2, df3, equal_var=False)
print(result)
if (result.pvalue > alpha):
     print('Не отвергаем нулевую гипотезу')
elif (result.statistic > 0):    
    print('Отвергаем нулевую гипотезу, критикам больше нравится PC ')
else:
    print('Отвергаем нулевую гипотезу, критикам больше нравится PS4 ')
```

Значение t-теста выше 0, значит первая выборка (PC), имеет среднее значение больше, чем конкурирующая выборка.

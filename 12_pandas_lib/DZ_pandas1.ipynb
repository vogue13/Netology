Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
highly_rated = ratings[ratings.rating == 5.0]['movieId'].value_counts().index[0]
```

```python
highly_rated 
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
movies['title'][movies.movieId == highly_rated].head()
```

Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
data.head()
```

```python
data['country'].sort_values().unique()
```

```python
data['category'].sort_values().unique()
```

```python
sorted_data = data[(data['country'].isin(['Latvia','Lithuania','Estonia']))& (data['category'].isin([4,12,21]))& 
                   ((data['year'] >= 2004)&(data['year'] <= 2010)) & (data['quantity'] >= 0)]
sorted_data
```

```python
sorted_data[['country', 'quantity']].groupby('country').sum().sort_values('quantity', ascending=False).head()
```
